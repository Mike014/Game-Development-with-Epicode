# 1) Tech Designer (Technical Designer)

* Figura ibrida tra **Game Design** e **Implementazione tecnica**.
* Trasforma idee in **sistemi funzionanti** tramite tool e scripting.
* Fa da **ponte** tra Game Designer e Programmatori, riducendo fraintendimenti nel passaggio **documento → prototipo** (soprattutto in team grandi).

# 2) Differenze tra ruoli

* **Game Designer**: regole, bilanciamento, flow; strumenti: documentazione, Excel, engine.
* **Tech Designer**: implementazione sistemi, prototipi; strumenti: Blueprints/Lua/Python/C#/Unreal/Unity.
* **Programmer**: codice, ottimizzazione, architettura; linguaggi: C++/C#, reti, engine custom.

# 3) Pitch di gioco — scopo, durata, stile

* Serve a far comprendere **idea ed esperienza** a consumer/publisher/stakeholder.
* **Durata**: ~5 minuti (max 7 con trailer/giocabile).
* **Stile**: discorsivo, **poche slide**, niente salti indietro.

# 4) Elementi chiave del Pitch

1. **Schermata iniziale**
   Splash art d’impatto: introduce subito tono e immaginario.
2. **Introduzione al gioco**
   Breve paragrafo con obiettivo e **mood** (anche frase ad effetto).
3. **Game Sheet**
   Scheda rapida: target, genere, piattaforme, tech, mercati, business model, prezzo.
4. **Game Structure**
   Concept, meccaniche base, esperienza; se narrativo, **sinossi**.
5. **Creative Mix / Video giocabile**
   Riferimenti a titoli/medie per chiarire l’idea; se c’è demo/trailer, mostrare materiale proprio.
6. **Game Loop**
   Cosa fa il giocatore nelle sessioni (diagramma delle attività ricorrenti).
7. **Game Pillars (identità non negoziabile)**

   * **Two Paths to Cross**: scelta Vendetta/Redenzione che altera gameplay, storia, ambienti.
   * **Heavenlands to Explore**: sei “terre della Luce” con poteri unici; ottenerli sconfiggendo o guadagnando la fiducia dell’Angelo.
   * **People to Meet, Stories to Live**: incontri significativi; aiutare gli NPC svela verità e porta alleati.
8. **Key Selling Points (vantaggi competitivi)**

   * **World Exploration**: 6 regioni diverse per meccaniche/creature/magia.
   * **The Light**: padronanza della magia angelica come sistema centrale.
   * **Puzzle ambientali**: enigmi risolti usando Luce + ambiente.
   * **Variable Story & Gameplay**: scelte con conseguenze reali.
   * **Party of Playable Characters**: il party cambia lo stile di gioco.
9. **Financial Projection / Timeline** *(cruciali per stakeholder)*

   * **Fondi necessari**: ~**550.000 €**
   * **Prezzo**: **14,99 €**
   * **Break-even**: ~**28.000** copie
   * **Previsione unità**: **500.000** (da ricerca competitor)
   * **Costo team (1 anno)**: **200.000 €**
   * **Timeline indicativa**:
     Pre-prod 2–4m → Prod 10–13m → Pre-Alpha 1–2m → Alpha 2–5m → Beta 2–3m → Testing 1–2m → Launch 1–2m
10. **Team**
    Presenta dimensione, competenze, contatti e profili chiave (es.: Game Director/Lead Designer/Producer; Concept/2D/3D Artist; UI/UX; Sound Designer/Composer; Lead 3D/Post).
