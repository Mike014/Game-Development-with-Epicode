# MDA: come usarlo davvero

* **Mechanics (M)** = regole, stati, risorse, azioni disponibili. Sono ciò che **progetti direttamente** (verbi di gioco, economie, collisioni, cooldown, stamina, IA, ecc.).
* **Dynamics (D)** = comportamenti **emergenti** quando i giocatori usano le meccaniche (kiting, meta-build, speedrunning, trading, griefing, aggro juggling). Nascono da incentivi, vincoli e feedback.
* **Aesthetics (A)** = l’esperienza percepita: tensione, catarsi, meraviglia, empowerment, umorismo, empatia. È ciò che **vuoi far sentire** al giocatore; musica, arte, narrativa e FX sono i veicoli.
* **Perché funziona**: progetta le **M** per produrre **D** che generino **A**. Se l’estetica desiderata non emerge, prima aggiusta incentivi/regole, non solo la presentazione.

---

# Mechanics (core vs supporto)

* Le **core/pillar mechanics** definiscono l’identità (es. “parry-counter” in un soulslike). Le meccaniche di supporto **non devono competere** col core (inventario, crafting leggero, hub).
* Principi operativi:

  * **Sottrazione**: la perfezione si raggiunge quando non c’è più nulla da togliere; elimina le meccaniche che non servono al core.
  * **Verbi chiari**: cosa può fare il giocatore? Saltare, schivare, hackerare, orientarsi con cue sonori, ecc.
  * **Leggi di sistema**: risorse (HP, mana, munizioni), trasformazioni (crafting), spazi d’azione (zone sicure/pericolose), tempo (cooldown, finestre di parry).
* Trappole comuni: “feature isole”, economie senza sink, timer che puniscono l’esplorazione, statistiche ridondanti.

---

# Dynamics (emergenza e “Meaningful Design”)

* Emergono da **incentivi + vincoli**: stamina lenta → ritmo cauto; crit legati al timing audio-visivo → gioco “ritmico”.
* **Meaningful Design**: le dinamiche **comunicano un valore/tema** (scarsità → tensione; eco-narrativa → ascolto attento/riconoscimento pattern).
* Come guidarle senza scriptarle:

  * Regola **rischio/ricompensa** (loot migliore in aree rumorose che attirano nemici).
  * Introduci **sinergie** (abilità + ambiente + audio cue).
  * Costruisci **loop di feedback** chiari: il gioco insegna col feedback (hit-stop, side-chain nel mix, vignetta, UI diegetica).
* Metriche utili: % uso abilità, tasso di accumulo risorse, tempo medio tra decisioni significative, varianza dei percorsi scelti.

---

# Aesthetics (obiettivi emotivi)

* Componenti: **musica**, **SFX**, **stile visivo**, **scrittura**, **ritmo/direzione**.
* Lavora con **obiettivi estetici** espliciti: “tensione sostenuta”, “scoperta”, “potenza”, “malinconia”. Ogni obiettivo ha **leve tecniche** (armonia, timbro, palette colori, densità HUD, illuminazione, curve di animazione).
* Regola: l’estetica **non** nasconde meccaniche deboli; **amplifica** quelle forti. Per “ansia”, restringi il FOV, riduci le opzioni sicure, sbilancia il mix sul diegetico, rimuovi musica non essenziale.

---

# CCC — Character, Camera, Controls

* **Character**: move-set, velocità, accelerazione, hit/hurt box, stamina; power fantasy e identità silhouette/sonica (passi, respiro, timbro arma).
* **Camera**: leggibilità e agency. Regole di lock-on, collision avoidance, offset dinamici; in 2D: parallasse, safe area; nell’horror: **occlusione controllata** e audio posizionale per “vedere con le orecchie”.
* **Controls**: latenza, mapping, coyote time, input buffering, deadzone, remapping e accessibilità. La frizione nei controlli distrugge dinamiche altrimenti valide.

---

# Game Loop (core e secondari)

* **Core loop**: ciclo breve che sostiene il 90% del tempo (Esplora → Ingaggia → Ricompensa → Micro-upgrade → repeat). Deve “pagare” in fretta.
* **Secondary loops**: progressioni più lente che **arricchiscono** il core (crafting, questline, mastery, relazioni con NPC). Offrono orizzonti di medio-lungo termine.
* Strumenti di tuning:

  * **Time-to-fun** (accendi il core entro 60–120 s).
  * **Durata del loop** e carico cognitivo (quanta UI/lettura tra due decisioni?).
  * **Sorgenti/Sink** dell’economia (evita inflazione delle ricompense).

---

# Come torna tutto insieme

1. **CCC** è il modo in cui i giocatori **toccano** le **Mechanics**.
2. Quelle meccaniche producono **Dynamics** osservabili.
3. Le **Aesthetics** confezionano e amplificano le dinamiche verso l’emozione target.
4. Il **Game Loop** è la **coreografia** che muove M-D-A nel tempo.
   → Se A non “atterra”, lavora a ritroso: aggiusta D (incentivi), poi M (regole/valori), poi CCC (responsività).




# SIMULACRA — Recap MDA/CCC 

**Tipo**: FMV “found phone” (telefono diegetico come UI).

## M — Mechanics (regole/azioni)

* Interfaccia telefono: chat, chiamate, galleria, note, browser.
* Azioni: leggere media, cercare **keyword**, sbloccare file/app, scegliere risposte, seguire link, comporre codici.
* Regole/Stati: password e condizioni di sblocco, **flag narrativi**, timer/eventi, rami multipli, info incomplete/glitch.

## D — Dynamics (comportamenti emergenti)

* **Indagine non lineare** (pattern-matching tra messaggi/foto/metadata).
* **Social engineering**: tono/timing nelle chat influenza fiducia e accessi.
* **Gestione del rischio**: link sospetti, credibilità delle fonti, forzare vs attendere.
* **Rilettura** degli indizi e backtracking quando emergono nuove piste.

## A — Aesthetics (esperienza/emozione)

* Urgenza/tensione (notifiche, chiamate improvvise, countdown).
* Voyeurismo/intrusione e **paranoia tecnologica**.
* Realismo diegetico (audio compresso, UI credibile, glitch controllati).

## CCC

* **Character**: giocatore-investigatore (co-protagonista “remota” = ragazza scomparsa e la sua rete).
* **Camera**: schermo del telefono (1ª persona); clip FMV come picchi emotivi.
* **Controls**: tap/scroll/typing, risposte a scelta, input time-critical; feedback audio/haptics.

## Game Loop

* **Core**: Leggi → **Inferisci** → **Scegli/Agisci** (chat/chiamata/sblocco) → **Conseguenza** (nuovi file/fiducia±/ramo) → Nuovo indizio → repeat.
* **Secondary**: progressione d’indagine (app/aree sbloccate), **relazioni/fiducia**, ricomposizione timeline, finali multipli.








## **Come si concatenano tra di loro?**
CCC è **l’interfaccia fisica**, MDA è **la logica/effetto**, il Game Loop è **il tempo** in cui tutto scorre. La catena completa, a livello operativo, è questa:

1. **Intenzione del giocatore → CCC (Controls/Camera/Character)**
   L’utente decide “voglio fare X”. I **Controls** traducono input in azioni; la **Camera** rende leggibile lo spazio; il **Character** esegue (move-set, timing, hitbox).

2. **CCC → Mechanics (M)**
   L’azione entra nel sistema: regole, risorse, stati, cooldown, gating, AI, economia. Qui avviene il **cambio di stato** misurabile.

3. **Mechanics → Dynamics (D)**
   Dall’incastro di regole, incentivi e vincoli emergono **comportamenti**: kiting, risk-taking, pathing, meta-build, social-engineering, ecc.

4. **Dynamics → Aesthetics (A)**
   Le dinamiche producono l’**esperienza percepita** (tensione, scoperta, potenza…). **Audio, UI, VFX, musica, scrittura** confezionano e amplificano quella sensazione.

5. **Aesthetics/Feedback → nuovo intento**
   Il feedback (sonoro/visivo/tattile) aggiorna il modello mentale del giocatore e innesca la **prossima decisione**.

6. **Game Loop = coreografia nel tempo**
   Questa catena 1→5 gira su più scale temporali:

   * **Core loop** (10–20 s): azione → conseguenza → ricompensa micro.
   * **Loop secondari** (2–10 min): progressi, sblocchi, relazioni.
   * **Meta loop** (ore): mastery, finali, metagame.
     Il loop assicura che ogni ciclo “paghi” e apra l’orizzonte del successivo.
