# **Fondamenti di Game Design**

### **Il mestiere: dal sistema all’esperienza**
Il game design è **progettazione di comportamenti e relazioni che provocano esperienze nel giocatore**. 
Il designer muove su un ciclo iterativo: **idea → prototipo → playtest → correzione**, con un principio guida: **ascoltare team**, **pubblico**, **gioco**, **cliente** e **sé stessi**. 
Il riferimento di cornice è l’**MDA**: 
- il designer progetta **“Mechanics”** (regole, risorse, spazi, azioni); 
- queste generano **“Dynamics”** (comportamenti emergenti in runtime); 
- il giocatore percepisce **“Aesthetics”** (tipi di divertimento/emozioni). 
La forza dell’MDA è **allineare l’intento del designer con la percezione del giocatore**, mantenendo il ciclo di tuning obiettivo.

---

### **Che cos’è “gioco”: i quattro mattoni operativi**

### 1) **Obiettivi**

Che cosa vuole ottenere il giocatore (e quando capisce di averlo ottenuto)?

* **Espliciti vs impliciti**: “Arriva al traguardo” (esplicito) vs “Gioca da ladro senza uccidere nessuno” (obiettivo autoimposto, implicito).
* **Orizzonte temporale**: breve (micro-task), medio (missione), lungo (campagna/metagame).
* **Criteri di successo/fallimento**: condizioni binarie o punteggi soglia; devono essere verificabili.
* **Perché contano**: forniscono direzione, priorità e ritmo. Senza obiettivi chiari, le scelte perdono significato.

### 2) **Regole**

Cosa è permesso/vietato e come evolve lo stato del sistema?

* **Affordance + vincoli**: definiscono quali azioni esistono e con che costi/limiti (risorse, cooldown, posizioni, linee di vista).
* **Stato e transizioni**: variabili (vite, oro, posizione, stati IA) e come cambiano.
* **Coerenza e completezza**: niente “buchi” dove un’azione non è definita; priorità/precedenze quando più regole si sovrappongono.

### Le 3 famiglie (Salen & Zimmerman, utile come lente):

* **Operational rules** (superficie): quello che “si fa” per giocare (tasti, turni, “peschi 1 carta”).
* **Constitutive rules** (struttura): la logica matematica/sistemica sottostante (probabilità, grafi di stato, economia).
* **Implicit rules** (sociali): etichetta e consuetudini (non barare, fair play).

### 3) **Feedback**

Come il mondo risponde alle azioni del giocatore?

* **Tempestività**: immediato (hit-spark, suono, shake) vs differito (risultato missione a fine livello).
* **Chiarezza**: cosa è successo, perché, e cosa cambia ora (stato, risorse, opportunità).
* **Canale**: diegetico (dentro il mondo: animazioni, UI in-world) vs non diegetico (HUD, numeri, log).
* **Bilanciamento**: feedback positivi/negativi e la loro “forza” (attenzione a loop positivi che esplodono/negativi che schiacciano).

### 4) **Decisioni**

Scelte **significative** quando comportano:

* **Trade-off reali** (rinunci a X per ottenere Y).
* **Conseguenze percepibili** (stato cambia, percorso si biforca).
* **Incertezza gestibile** (informazioni parziali, rischio calcolabile).
* **Informazione sufficiente** (telegraphing) per sentire agency, non lotteria cieca.
* **Tempismo** (quanto è reversibile? c’è pressione di tempo?).

## Esempio didattico: “Tre Forzieri”

Un micro-gioco che puoi spiegare in 60 secondi e usare per mostrare i 4 mattoni.

### Pitch

In 6 turni devi accumulare **≥ 20 gemme** senza perdere tutte le **3 vite**. Ogni turno scegli **1** dei **3 forzieri**.

### Obiettivi

* **Primario**: raggiungere 20 gemme entro 6 turni.
* **Secondario**: non finire le 3 vite.
* **Chiusure**: vinci se Gemme ≥ 20; perdi se Vite = 0 o finiti i turni senza soglia raggiunta.

### Regole

* Setup di turno: si randomizzano 3 forzieri (2 con **+5 gemme**, 1 **Trappola**: −1 vita).
* Azioni del giocatore per turno:

  1. **Apri** un forziere (obbligatorio se non salti).
  2. **Opzionale**: **Occhiata** (peek) a 1 forziere **spendendo 2 gemme** (rivela il contenuto di quel forziere solo a te per questo turno).
  3. **Opzionale**: **Salta** il turno per guadagnare **1 Gettone Indizio** (max 2) che sostituisce in futuro la spesa di 2 gemme per un’occhiata.
* Fine turno: aggiorna gemme/vite; decrementa turni rimanenti.

#### Tassonomia delle regole (stesso gioco, tre strati)

* **Operational**: “Ogni turno scegli tra Apri/Occhiata/Salta; aprire consuma il turno; occhiata costa 2 gemme o 1 gettone; 6 turni totali; forziere trappola toglie 1 vita”.
* **Constitutive**: distribuzione 2/3/1 dei contenuti, cap dei gettoni = 2, soglia vittoria = 20, funzione ricompensa = +5, costo occhiata = 2, probabilità e stato (vite/gemme/turni).
* **Implicit**: non segnarti l’ordine dei forzieri oltre ciò che il gioco consente; niente “reroll” manuale; fair play.

### Feedback

* **Immediato**: apertura forziere → effetto visivo/sonoro differente (brillio per +5, colpo sordo + schermo che trema per trappola), HUD aggiorna Gemme/Vite.
* **Informativo**: occhiata mostra icona sopra il forziere “sicuro” o “trappola”; messaggio “2 gemme spese / 1 gettone usato”.
* **Differito**: a fine turno un riepilogo rapido; a fine partita schermata esito con breakdown (quante occhiate, quante trappole, media gemme/turno).

### Decisioni

* **Scelta di rischio**: aprire subito (2/3 di guadagnare 5, 1/3 di perdere 1 vita) vs spendere per ridurre incertezza.
* **Gestione risorse**: spendere gemme ora rallenta il raggiungimento della soglia, ma evita perdite di vite (che a medio termine possono azzerare la run).
* **Tempo vs informazione**: saltare per ottenere gettoni riduce i turni utili, ma renderà più sicure le scelte future.
* **Adattività**: se dopo 3 turni hai 0/3 vite perse e 10 gemme, la tua policy può cambiare (più aggressiva); se sei a 1 vita, conviene investire in occhiate.

### **Che cos’è “gioco”**: i quattro mattoni operativi

**[MDA](https://users.cs.northwestern.edu/~hunicke/MDA.pdf)** in pratica: collegare **leve (M)** a **comportamenti (D)** e a **emozioni (A)**
- **Mechanics** → **Dynamics**: vincoli, risorse, spazi e tempi generano pattern (controllo dell’area, rischio/ricompensa, loop economici, meta).
- **Dynamics** → **Aesthetics**: l’insieme produce “tipi di divertimento” (sfida, fantasia, narrativa, scoperta, espressione, socialità, sensazione, passatempo). Usare una tavolozza di “8 Kinds of Fun” aiuta a esplicitare l’obiettivo estetico di ogni feature. 

Un gioco è un **sistema che produce comportamento**. Per progettarlo e studiarlo, MDA scompone il sistema in **tre strati collegati** causa→effetto:

```
( Prospettiva del designer )             ( Prospettiva del giocatore )
Mechanics  →  Dynamics  →  Aesthetics    Aesthetics  →  Dynamics  →  Mechanics
  (regole)     (comportamenti)  (emozioni)    (attese)    (ciò che accade) (controlli visibili)
```

* **Mechanics**: dati, regole, algoritmi, azioni, risorse, RNG, AI, livelli.
* **Dynamics**: ciò che emerge in runtime dall’interazione tra meccaniche e input (press. del tempo, economia, meta, cooperazione, bluff, stealth…).
* **Aesthetics**: le **esperienze**/emozioni desiderate nel giocatore.

> Pensare il gioco come **artefatto** (non “media” lineare) aiuta a ragionare su cause tecniche → effetti esperienziali.

---

#### Lessico delle Estetiche (8 tipi di “fun”)

1. **Sensation** (sensazioni)
2. **Fantasy** (finzione/ruolo)
3. **Narrative** (dramma/storia)
4. **Challenge** (sfida)
5. **Fellowship** (socialità/cooperazione)
6. **Discovery** (scoperta/esplorazione)
7. **Expression** (personalizzazione/auto-espressione)
8. **Submission** (passatempo/flow tranquillo)

> Ogni gioco combina più estetiche, con pesi diversi (es. *Quake*: Challenge, Sensation, Competition; *The Sims*: Discovery, Fantasy, Expression, Narrative).

---

#### Dinamiche tipiche (come ottenere le estetiche)

* **Challenge** ← timer, avversari, risorse scarse, rischio/ricompensa.
* **Fellowship** ← obiettivi condivisi, ruoli complementari, info parziali tra compagni.
* **Expression** ← sistemi di build/crafting, editor, skill tree, cosmetici.
* **Narrative/Tensione** ← cicli di escalation → climax → rilascio, gating di info, foreshadowing.
* **Discovery** ← mappe non note, segreti, progressione di conoscenza.
* **Sensation** ← feedback audiovisivo ricco, fisica tattile.
* **Submission** ← loop ripetibili, attrito basso, progressi piccoli ma costanti.

#### Schema operativo “4 mattoni” (riuso rapido)

```
OBIETTIVI
- Cosa devo ottenere? In quanto tempo? Criteri di successo/fallimento?

REGOLE (Operational / Constitutive / Implicit)
- Cosa posso fare? Costi/limiti? Stato e transizioni? Priorità conflitti?
- Strato costitutivo: formule, probabilità, economie, FSM, AI.
- Strato implicito: fair play, norme sociali (multiplayer/co-op).

FEEDBACK
- Immediato (hit, UI, suono), differito (end screen); chiarezza “cosa/Perché/Adesso cosa”.
- Canali diegetici/non diegetici; forza del feedback (evita saturazione).

DECISIONI
- Trade-off reali? Info sufficiente ma non totale? Rischio calcolabile? Reversibilità?
- Pressione di tempo o di risorsa? Impatto percepibile sullo stato?
```

---

### In sintesi

* MDA ti obbliga a **partire dall’esperienza** e a scegliere meccaniche **al servizio** di dinamiche mirate.
* La qualità arriva dal **tuning iterativo**: spostare numeri/regole finché le dinamiche osservate realizzano l’estetica desiderata, mantenendo tensione e agency.
* Parlare con il vocabolario M-D-A rende discussione, debugging e ricerca **oggettivi**: puoi dire “questa meccanica introduce un loop positivo che rompe la *Challenge* prevista” invece di “non è divertente”.

### **Tassonomia delle Mechanics nelle slide**

* **Time**: tempo nei giochi è artificiale. Può essere **Discreto**, o **Continuo**, cioè a turni o in tempo reale. Il tempo può essere circoscritto: **Clock**, o può essere un avversario **Races**.
* **Space**: discreto (celle/griglie) vs continuo; numero di dimensioni; aree connesse/non connesse. Scelte topologiche influenzano navigazione, controllo e leggibilità.
* **Objects**: qualunque cosa manipolabile. Ogni oggetto ha **Attributes** con **Current State**; molti attributi chiave sono modellati con **Finite-State Machines** (es. routine dei fantasmi in **Pac-Man**: stati come “chase/scatter/frightened/eyes” e transizioni su timer, power-up, collisioni). I fantasmi di Pac-Man hanno comportamenti apparentemente complessi generati da AI sorprendentemente semplici, Ogni fantasma ha una "tile target" (casella obiettivo) che cerca di raggiungere. I fantasmi pianificano solo un passo alla volta - quando entrano in una nuova casella, decidono dove girare nella prossima basandosi su quale direzione li avvicina al target. Non possono mai invertire direzione, in caselle con solo due uscite, continuano sempre dritto, le tile target possono essere irraggiungibili.
* **Actions**: di base (sempre disponibili) vs **strategiche** (acquistano senso rispetto all’obiettivo). Progettare costi, condizioni e leggibilità dell’azione.
* **Rules**: strato fondamentale. Distinguere regole **operative** (come si gioca), **costitutive** (logica interna), **implicite** (etichetta/sportività) evita ambiguità e bug di design. 
* **(Real) Skill**: abilità fisiche/mentali/sociali del **giocatore reale**. Il design efficace allena la maestria con curva di difficoltà, feedback e scaffolding.
* **Chance**: il caso introduce **incertezza** utile; crea tensione, opportunità di reazione e momenti “clutch”. Troppa alea degrada in casualità opaca; poca alea appiattisce la rigiocabilità. Bilanciare alea e abilità è un asse progettuale centrale. 

### **Implicazioni progettuali chiave (da usare nel GDD)**

* **Agency leggibile**: il giocatore deve capire “cosa posso fare ora e perché conta” (azioni, affordance, feedback).
* **Tensione regolata**: calibrare **uncertainty** (chance, informazione nascosta, complessità situazionale) perché sostenga sfida e flow senza frustrare. 
* **Loop e economie**: definire loop core (azioni→feedback→ricompense→nuove possibilità) e i **limiti** (cooldown, risorse) che ne danno forma.
* **Stati e transizioni**: mappare FSM per i sistemi con switching (IA, mission states, UI). Riduce ambiguità e bug; facilita tuning. 
* **Allineamento M→A**: ogni meccanica dovrebbe dichiarare quale **Aesthetic** intende servire (es. “risk-reward” per **Challenge/Discovery**). 
* **Regole trasparenti**: separare “regole per i giocatori” (manuale/UI) da “regole costitutive” (motore) e “implicite” (etichetta) previene conflitti.
