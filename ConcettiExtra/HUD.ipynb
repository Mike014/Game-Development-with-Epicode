## HUD (Head-Up Display) nel Game Development

### Definizione

Un **HUD (Head-Up Display)** è l’insieme di **informazioni visive sovrapposte alla schermata di gioco** che comunicano al giocatore lo **stato del personaggio e del gameplay in tempo reale**, senza interrompere l’azione.

Il termine deriva dall’aviazione, dove gli strumenti sono proiettati direttamente nel campo visivo del pilota. Nei videogiochi il concetto è analogo: informare **senza distrarre**.

## Funzione principale

Il compito di un HUD è rispondere a una domanda chiave:

> **Quali informazioni servono al giocatore adesso per prendere decisioni corrette?**

Per questo l’HUD è:

* sempre (o quasi) visibile
* aggiornato in tempo reale
* progettato per una lettura immediata

## Informazioni tipiche di un HUD

Un HUD può includere:

* **Stato del personaggio** (vita, stamina, mana)
* **Risorse** (munizioni, cariche, cooldown)
* **Orientamento** (minimappa, bussola)
* **Feedback di combattimento** (crosshair, hit marker, indicatori di danno)
* **Obiettivi e notifiche rapide**

## HUD vs UI

Nel game development si distingue spesso tra:

### HUD

* Parte del **gameplay attivo**
* Sempre visibile o contestuale
* Ottimizzato per l’azione rapida

### UI (User Interface)

* Menu, inventario, pause
* Interazioni più lente e riflessive
* Non direttamente legata all’azione in tempo reale

## Principi di design di un buon HUD

Un HUD efficace deve essere:

1. **Non invasivo**
   Non coprire elementi importanti della scena.
2. **Immediato**
   Comprensibile in una frazione di secondo.
3. **Contestuale**
   Mostrare solo le informazioni rilevanti.
4. **Coerente**
   Allineato allo stile e al tono del gioco.

## Tipologie di HUD

* **Statico**: sempre visibile (es. barre di stato)
* **Dinamico**: appare solo quando necessario
* **Diegetico**: integrato nel mondo di gioco
  (es. indicatori sul casco, display sull’arma)

## HUD in Unity (concetto tecnico)

In Unity, un HUD è generalmente:

* un **Canvas** in modalità **Screen Space**
* composto da elementi UI (Image, Text, Slider, ecc.)
* aggiornato via script leggendo lo stato del gameplay

## Sintesi finale

> Un HUD è l’interfaccia informativa in tempo reale che supporta il gameplay, fornendo al giocatore le informazioni essenziali senza interrompere l’esperienza di gioco.
