**Introduzione al controllo del personaggio**

Il **Character Controller** è un componente progettato per gestire il movimento del personaggio nei giochi in prima o terza persona, dove si privilegia un controllo reattivo e diretto piuttosto che un realismo fisico completo.

**Caratteristiche principali:**
*   Utilizza un **Collider a forma di capsula** mantenuto sempre verticale.
*   **Non richiede un Rigidbody** e ignora gli effetti fisici realistici come l'inerzia e la spinta delle forze esterne.
*   Si muove in base a comandi diretti (es. velocità e direzione impostate da script), ma viene fermato dalle collisioni con oggetti statici (pareti, pavimenti) e può salire gradini o pendii entro limiti definiti.
*   Può spingere oggetti con Rigidbody, ma **non viene spostato dalle collisioni** con essi.

**Proprietà fondamentali per la regolazione:**
*   **Slope Limit:** Angolo massimo dei pendii che il personaggio può salire.
*   **Step Offset:** Altezza massima dei gradini superabili.
*   **Skin Width:** Profondità di penetrazione consentita tra collider. **Critica per evitare che il personaggio si incastri**. Si consiglia un valore > 0.01 e superiore al 10% del raggio.
*   **Radius & Height:** Definiscono le dimensioni della capsula collider. Per un personaggio umanoide si consiglia un'altezza di circa 2 metri.
*   **Min Move Distance:** Di solito lasciato a 0 per evitare scatti.

**Scopo e uso consigliato:**
È la soluzione ideale per controlli "arcade", dove il personaggio si muove, ferma e gira immediatamente (come nei classici sparatutto in prima persona). Se invece il personaggio deve reagire in modo realistico a forze fisiche (es. essere spinto), è più adatto un **Rigidbody**.

**Suggerimenti:**
1.  Se il personaggio si incastra, aumentare la **Skin Width**.
2.  La logica per spingere o interagire con oggetti fisici deve essere implementata manualmente tramite script (usando `OnControllerColliderHit()`).
3.  Modificare le proprietà in fase di esecuzione ricrea il controller, il che può interrompere temporaneamente il rilevamento delle collisioni.
