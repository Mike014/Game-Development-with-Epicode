### Spiegazione concettuale del Day & Night Cycle

Il **Day & Night Cycle** è un sistema che simula il passare del tempo in un mondo di gioco trasformando la luce, i colori e l’illuminazione globale della scena. Non è legato a un genere specifico: è una **macchina del tempo visiva** riutilizzabile in qualsiasi progetto.

---

## 1. Il tempo come variabile astratta (0 → 1)

Il cuore del sistema non sono le ore reali, ma una **variabile normalizzata**:

* `time = 0` → mezzanotte
* `time = 0.5` → mezzogiorno
* `time = 1` → fine giornata (riparte da 0)

Questo rende il sistema:

* indipendente dalla durata reale del giorno
* facile da mappare su curve, colori e rotazioni

Il tempo avanza in modo continuo e ciclico, come una **funzione periodica**.

---

## 2. Sole e Luna come due stati opposti

Sole e Luna non sono due sistemi separati, ma **due facce dello stesso ciclo**:

* quando il Sole sale, la Luna scende
* quando il Sole è al massimo, la Luna è “sotto il mondo”
* non sono mai dominanti insieme

Concettualmente:

> la Luna è il Sole sfasato di mezzo ciclo

Questo garantisce coerenza naturale senza logiche complesse.

---

## 3. La rotazione rappresenta il tempo, non il movimento reale

Il Sole **non si muove nello spazio**, ma ruota attorno al mondo:

* la rotazione sull’asse X è una **metafora visiva del tempo**
* mezzogiorno è un riferimento fisso (es. 90°)
* l’angolo deriva direttamente da `time`

In pratica:

> l’orientamento della luce *è* l’orologio del mondo

---

## 4. Curve e gradienti: il comportamento nasce dai dati, non dal codice

Il sistema non usa `if (giorno)` / `if (notte)`:

* **Animation Curve** → controlla *quanto* è intensa la luce in ogni momento
* **Gradient** → controlla *che colore* ha la luce in ogni momento

Questo significa che:

* alba e tramonto sono **zone di transizione**, non eventi istantanei
* il comportamento visivo è disegnato, non programmato

Concetto chiave:

> il codice esegue, le curve decidono

---

## 5. Accendere e spegnere le luci è una scelta fisica, non estetica

Quando l’intensità del Sole arriva a zero:

* il Sole viene **disattivato**, non solo oscurato

Questo evita:

* artefatti di luce
* illuminazioni innaturali dal basso

È una scelta concettuale importante:

> se una fonte luminosa non emette energia, non dovrebbe esistere nella scena

---

## 6. L’illuminazione ambientale racconta il “clima” del mondo

Anche senza Sole o Luna, il mondo non dovrebbe brillare:

* **Ambient Intensity** → quanta luce “residua” esiste
* **Reflection Intensity** → quanto l’ambiente riflette il cielo

Di giorno:

* valori alti → mondo leggibile e luminoso

Di notte:

* valori bassi → buio credibile, atmosfera più drammatica

Questo separa:

* **luce diretta** (Sole/Luna)
* **luce indiretta** (cielo, atmosfera)

---

## 7. Il sistema è una funzione continua, non una sequenza di stati

Concettualmente il Day/Night Cycle non è:

* Giorno → Notte → Giorno

Ma:

* una **funzione continua del tempo**

Ogni elemento (rotazione, colore, intensità, ambiente) è:

```
valore = f(time)
```

Questo lo rende:

* prevedibile
* estendibile
* perfetto per collegare gameplay, AI, eventi, audio
