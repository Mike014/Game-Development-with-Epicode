# Panoramica dei Formati Audio e della Spazializzazione

Una guida per comprendere i principali formati audio, dal mono all'audio immersivo, con spiegazioni semplici su come funzionano e dove vengono utilizzati.

## Classificazione per Numero di Canali

### 1. Mono
* **Descrizione**: Un singolo segnale audio ripetuto identico su tutti gli altoparlanti.
* **Percezione**: Il suono viene percepito come proveniente da un unico punto, senza alcuna spazialità direzionale.
* **Utilizzo Principale**: Voci (telefoni, podcast), altoparlanti singoli, sistemi di comunicazione e annunci.

### 2. Stereo
* **Descrizione**: Due canali audio indipendenti: Destro (R) e Sinistro (L).
* **Percezione**: Crea un'immagine sonora orizzontale, permettendo di distribuire suoni tra i due lati. Offre una sensazione di larghezza.
* **Utilizzo Principale**: Musica in streaming, televisione, radio, videogiochi e computer.

### 3. Surround (es. 5.1, 7.1)
* **Descrizione**: Utilizza più canali fisici (es. 5.1 = 5 canali principali + 1 subwoofer) posizionati attorno all'ascoltatore.
* **Percezione**: Il suono avvolge l'ascoltatore, provenendo anche dai lati e da dietro, per un'esperienza cinematografica.
* **Utilizzo Principale**: Cinema, home theater, film in Blu-ray, alcuni videogiochi per console e PC.

## Formati Audio Immersivi e Spaziali

### 4. Binaurale
* **Descrizione**: Tecnica di registrazione che usa due microfoni posizionati dentro un manichino con orecchie, per simulare l'ascolto umano.
* **Percezione**: Creata esclusivamente per le cuffie, produce un'illusione 3D ultra-realistica: si percepiscono suoni sopra, sotto, dietro e a distanze precise.
* **Nota Tecnica**: Sfrutta le HRTF (Funzioni di Trasferimento Relative alla Testa) per simulare come la testa e i padiglioni auricolari modificano il suono. È un'evoluzione del Stereo per cuffie.

### 5. Ambisonics
* **Descrizione**: Formato di registrazione e rappresentazione che cattura l'intero campo sonoro sferico a 360°, indipendente dal sistema di riproduzione.
* **Percezione**: Totalmente immersivo. Il file Ambisonics (B-Format) deve essere decodificato per il sistema di ascolto desiderato (cuffie, stereo, surround, array di altoparlanti).
* **Utilizzo Principale**: Realtà Virtuale (VR), video a 360°, contenuti interattivi e installazioni d'arte sonora.

## Sistemi di Audio ad Oggetti

### 6. Dolby Atmos (e simili: DTS:X)
* **Descrizione**: Sistema "object-based". Oltre ai canali tradizionali, i singoli suoni (oggetti audio) possono essere posizionati e mossi con precisione in uno spazio 3D, incluso l'altezza.
* **Percezione**: Audio tridimensionale con grande precisione. I suoni possono provenire da qualsiasi punto, anche dal soffitto (es. pioggia, aerei).
* **Flessibilità**: Può essere adattato (renderizzato) per sistemi diversi: da un home theater completo con altoparlanti sul soffitto a una soundbar, fino a un paio di cuffie (Atmos Binaurale).
* **Utilizzo Principale**: Cinema di ultima generazione, home theater premium, streaming (Netflix, Disney+), musica (Apple Music, Tidal) e videogiochi moderni.

## Tabella Riassuntiva e Confronti

| Formato | Cosa Cattura / Crea | Percezione Ideale | Utilizzo Tipico |
| :--- | :--- | :--- | :--- |
| **Mono** | Una singola fonte sonora. | Nessuna direzione. | Podcast, annunci, telefonia. |
| **Stereo** | Un'immagine sonora orizzontale (L/R). | Diffusione laterale. | Musica, film, TV generalista. |
| **Surround** | Un campo sonoro multi-direzionale fisso. | Avvolgimento a livello dell'orecchio. | Cinema, home theater classico. |
| **Binaurale** | Come l'orecchio umano sente nello spazio. | Cuffie. Realismo 3D estremo. | ASMR, tour virtuali, esperienze audio per cuffie. |
| **Ambisonics** | L'intera sfera sonora 360°. | Dipende dalla decodifica (cuffie o speaker). | VR, audio immersivo interattivo. |
| **Dolby Atmos** | Oggetti audio in uno spazio 3D + canali. | Sistema multi-altoparlante o cuffie. | Cinema, streaming premium, giochi next-gen. |

### Confronti Chiave
* **Stereo vs Binaurale**: Entrambi usano 2 canali, ma il binaurale è progettato per le cuffie con HRTF per un suono 3D realistico, mentre lo stereo è ottimizzato per altoparlanti.
* **Binaurale vs Ambisonics**: Il binaurale è un prodotto finale per cuffie. L'Ambisonics è una materia prima flessibile che può essere decodificata in binaurale, stereo, surround, ecc.
* **Surround 7.1 vs Dolby Atmos**: Il 7.1 ha 7 canali fissi a livello dell'orecchio. L'Atmos usa canali + oggetti mobili in 3D, aggiungendo la dimensione dell'altezza e una precisione dinamica superiore.

## Frasi da Ricordare
* **Mono** = un punto.
* **Stereo** = una linea (sinistra/destra).
* **Surround** = un cerchio attorno a te.
* **Binaurale** = il mondo reale nelle tue cuffie.
* **Ambisonics** = una sfera sonora da decodificare.
* **Dolby Atmos** = oggetti audio posizionabili ovunque nello spazio 3D.
