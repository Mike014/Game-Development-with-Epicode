## **AudioSettings** 

La **AudioSettings class** gestisce le **impostazioni audio globali** e permette di **riconfigurare il sistema audio a runtime**.

## Cosa controlla

* **Speaker Mode** (stereo, surround, ecc.)
* **Sample Rate** (se supportato dalla piattaforma)
* **DSP Buffer Size**
* **Real / Virtual Voice Count**

## Project Settings vs Runtime

* **Project Settings â†’ Audio**

  * Definiscono lo **stato iniziale** del gioco
  * Valide per Editor e Build
* **AudioSettings API**

  * Cambia le impostazioni **solo a runtime**
  * Alla fine del Play, tutto torna come nei Project Settings

## API principali

| API                           | Funzione                           |
| ----------------------------- | ---------------------------------- |
| `GetConfiguration()`          | Legge configurazione audio attuale |
| `Reset(config)`               | Applica nuova configurazione audio |
| `OnAudioConfigurationChanged` | Callback su cambio device/config   |

ðŸ‘‰ Queste API **rimpiazzano** i vecchi setter diretti (`outputSampleRate`, `speakerMode`, ecc.).

## Cambi di configurazione (importante)

Le modifiche possono avvenire:

* dal **menu opzioni audio** del gioco
* da **cambio hardware** (cuffie, HDMI, device audio)

âš ï¸ **Effetto collaterale critico**:

## Concetto chiave (da ricordare)

### Frase da esame
