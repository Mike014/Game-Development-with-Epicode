In Unity, la proprietà **Scale** del componente Transform non rappresenta una misura assoluta in metri o unità, ma è un **moltiplicatore** relativo alle dimensioni originali della mesh dell'oggetto. Questo spiega perché oggetti diversi con lo stesso valore di Scale (es. 1, 1, 1) mostrano dimensioni reali differenti nel mondo di gioco.

### Dimensioni Standard dei Primitivi (Scale 1, 1, 1)

Secondo la documentazione ufficiale di Unity, i "magic numbers" delle dimensioni base dei modelli primitivi sono i seguenti:

| Oggetto Primitivo | Dimensioni Reali (in Unity Units) | Note Documentazione |
| --- | --- | --- |
| **Cube** | $1 \times 1 \times 1$ | Ogni lato misura esattamente 1 unità. |
| **Sphere** | Diametro 1 (Raggio 0.5) | Occupa lo spazio di un cubo . |
| **Quad** | $1 \times 1$ | Un piano sottile composto da soli 2 triangoli. |
| **Capsule** | Diametro 1, Altezza 2 | Composta da un cilindro con calotte sferiche. |
| **Cylinder** | Diametro 1, Altezza 2 | Il corpo centrale è alto 2 unità. |
| **Plane** | $10 \times 10$ | **Attenzione:** Un piano a scala 1 è 10 volte più grande di un cubo. |

### Ricerca e Fondamenti Tecnici

Dall'analisi della documentazione e delle best practices per sviluppatori (Unity Manual & TechArtHub), emergono i seguenti punti chiave:

1. **Convenzione 1 Unit = 1 Meter:** Unity Technologies ha calibrato i sistemi di fisica (Nvidia PhysX), l'illuminazione e le ombre assumendo che 1 unità corrisponda a 1 metro reale. Oggetti troppo grandi o troppo piccoli rispetto a questa scala possono causare artefatti visivi o comportamenti fisici instabili.
2. **Origine dei Modelli:** Per i modelli importati da software esterni (Blender, Maya, 3ds Max), la dimensione reale a Scale 1 dipende dalle impostazioni di esportazione dell'artista. Unity applica un "Scale Factor" nell'Inspector dell'asset per tentare di normalizzare queste differenze.
3. **Il "Plane Trap":** Il Plane è stato progettato storicamente come una mesh  suddivisa in 200 triangoli per supportare meglio la deformazione e la ricezione delle ombre. Per piani semplici , la documentazione suggerisce l'uso del **Quad**.
