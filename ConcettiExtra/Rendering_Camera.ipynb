## **Cosa è:**

Questa è la **configurazione del metodo di rendering della camera** - cioè come Unity disegna gli oggetti nella scena.

---

## **Le tre opzioni:**

### **1️⃣ Use Graphics Settings** (Consigliato - di default)
```
Usa le impostazioni globali definite in Graphics Settings
```
- La camera segue le impostazioni generali del progetto
- **Pro:** Coerente in tutta l'app
- **Contro:** Meno controllo personalizzato

---

### **2️⃣ Use Forward** 
```
Render all objects with one pass per light (Un passaggio per ogni luce)
```
- Disegna ogni oggetto **una volta per ogni luce** nella scena
- Se hai 3 luci → ogni oggetto viene disegnato 3 volte
- **Pro:** Buono per poche luci (1-3)
- **Contro:** Molto costoso se hai molte luci (10+)

---

### **3️⃣ Use Deferred**
```
Draw all objects once without lighting, then draw the lighting at the end
```
- Disegna **prima tutti gli oggetti senza illuminazione**
- **Poi calcola l'illuminazione una sola volta** alla fine
- Se hai 3 luci → ogni oggetto viene disegnato **1 sola volta** + calcolo luci
- **Pro:** Ottimo per MOLTE luci (20+)
- **Contro:** Consuma più memoria, effetti speciali più complicati

---

### **4️⃣ Use Legacy Vertex Lit**
```
Render all lights in a single pass, calculated at vertices
```
- Illuminazione calcolata **sui vertici** (angoli del modello), non sui pixel
- **Pro:** Molto veloce, stile vintage/retrò
- **Contro:** Qualità grafica inferiore

---

## **QUALE SCEGLIERE?**

| Situazione | Scelta |
|-----------|--------|
| **Gioco 3D moderno** | Use Graphics Settings ✓ |
| **Poche luci (1-3)** | Use Forward |
| **MOLTE luci (10+)** | Use Deferred |
| **Gioco retrò/low-poly** | Use Legacy Vertex Lit |
