## Esempio Pratico

```csharp
using UnityEngine;

public class Esempio : MonoBehaviour
{
    [ContextMenu("Debug Position")]
    public void PrintPosition()
    {
        Debug.Log($"Posizione: {transform.position}");
    }

    [ContextMenu("Reset Transform")]
    public void ResetTransform()
    {
        transform.position = Vector3.zero;
        transform.rotation = Quaternion.identity;
    }
}
```

## Regole Importanti

| Regola | Esempio |
|--------|---------|
| **Deve essere public** | `private void Metodo()` â†’ Non appare |
| **Niente parametri** | `public void Metodo(int x)` â†’ Errore |
| **Funziona in Editor** | Sia Play che Edit mode |
| **Non serve PlayMode** | Puoi usarlo senza premere Play |

---

## Quando Usarlo

- **Test e Debug**: Trigger rapidi senza UI  
- **Workflow veloce**: Genera mesh, resetta livello, esporta dati  
- **One-off actions**: Operazioni che fai una volta durante sviluppo  

- **Non usare** per gameplay (serve un sistema input vero)
