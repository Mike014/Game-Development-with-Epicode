# **Particle System (Sistema Particellare)**

### **Cosa sono gli Effetti Particellari (Le fondamenta)**

Il **Particle System** come una bomboletta spray intelligente. Invece di dipingere, "spruzza" nel mondo di gioco centinaia di piccole immagini 2D (o piccoli mesh 3D) chiamate **particelle**. Ogni particella ha una sua piccola "vita": nasce, si muove secondo delle regole, invecchia e poi muore scomparendo.

Ogni Particle System è un **Componente** associato a un GameObject.
Il GameObject è solo un contenitore vuoto con una posizione nello spazio (Transform); è il componente Particle System a dargli il "potere" di emettere fiamme.

Il Particle System di Unity è estremamente complesso, quindi per renderlo gestibile è stato diviso in Moduli. È come avere un pannello di controllo modulare in cui accendi solo i circuiti che ti servono.

- Modulo Base: È sempre attivo e definisce le regole universali della "vita" della particella: quanto vive (Lifetime), quanto è veloce all'inizio (Start Speed), quanto è grande (Start Size).
- Moduli Opzionali (i comportamenti): Li attivi con la spunta. Le slide ne citano alcuni fondamentali:
  - Emission: È il rubinetto. Quante particelle devono uscire al secondo? Escono in un flusso continuo o scoppiano tutte insieme (Burst)?
  - Shape: Da quale forma geometrica vengono emesse? Se fai una torcia userai un Cono stretto; se fai la pioggia userai un Box molto largo sopra la telecamera.
  - Color over Lifetime: Come evolve il colore nel tempo? (Il fuoco, ad esempio, nasce giallo/bianco al centro, diventa arancione e muore grigio scuro sfumando nel fumo).

**La Best Practice**: La slide dà un consiglio d'oro: "partire da un ParticleSystem già esistente e modificarlo". Nel game dev, fare reverse engineering di sistemi funzionanti è il modo più rapido per imparare come i vari moduli interagiscono tra loro.

**Sistemi Complessi (Composizione e Gerarchia)**
Questa è la slide concettualmente più importante per capire come ragionare in Unity.

- **Il Limite:** Un singolo componente Particle System può gestire un solo tipo di comportamento e texture per volta.
- **La Soluzione:** Se vuoi creare un falò realistico, un solo tipo di particella non basta. Il falò è fatto da fiamme vive, scintille che volano verso l'alto, fumo denso, e la luce vera e propria che illumina l'ambiente.
- **L'Approccio Architetturale:** Sfruttiamo la Gerarchia di Unity. Crei un GameObject "Padre" (che magari gestisce le fiamme base), e al suo interno inserisci altri GameObject "Figli", ciascuno col suo Particle System dedicato (uno per le scintille FireEmbers, e magari un componente Light standard di Unity per l'illuminazione). Spostando il Padre, si muoverà l'intero "pacchetto" coerentemente.
