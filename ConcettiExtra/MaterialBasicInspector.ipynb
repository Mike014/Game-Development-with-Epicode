### 1. L'Intestazione: Il Shader

In cima a tutto c'è un menu a tendina (di solito c'è scritto `Standard`).

* **Cos'è:** È il "cervello" matematico che decide come disegnare l'oggetto.
* **Importante:** Se cambi questo (es. su *Unlit* o *Mobile*), tutte le opzioni sotto cambieranno. Per ora, lascia sempre su **Standard** (o *Universal Render Pipeline/Lit* se usi URP).

### 2. Main Maps (Le Proprietà Principali)

Questa è la sezione dove passi il 90% del tempo. Definisce l'aspetto fisico.

#### **A. Albedo (Il Colore Base)**

Vedi un piccolo quadrato (per la texture) e un rettangolo colorato.

* **Funzione:** Rappresenta il colore "piatto" dell'oggetto, senza luci o ombre. È la "vernice".
* **Come si usa:**
* **Colore:** Clicca sul rettangolo per scegliere una tinta (es. Rosso Ferrari).
* **Texture:** Trascina un'immagine (es. una foto di legno) nel quadratino piccolo. L'oggetto verrà "avvolto" da quella foto.

#### **B. Metallic & Smoothness (Il "Feel" della superficie)**

Questi due slider lavorano insieme e sono il cuore del realismo.

* **Metallic (0 a 1):**
* **0 (Non Metallo):** Plastica, Legno, Tessuto, Pelle, Cemento.
* **1 (Metallo):** Oro, Ferro, Rame, Cromo.
* *Nota:* Nella realtà, gli oggetti sono quasi sempre o 0 o 1. I valori intermedi si usano solo per metalli sporchi o arrugginiti.


* **Smoothness (0 a 1):** (Spesso è lo slider sotto Metallic)
* **0 (Ruvido):** Una palla da tennis, asfalto, gesso. La luce viene "sparpagliata" e non c'è riflesso nitido.
* **1 (Liscio/Lucido):** Uno specchio, vetro, palla da biliardo bagnata. Il riflesso è nitido e perfetto.

| Metallic | Smoothness | Risultato Visivo |
| --- | --- | --- |
| 0 | 0 | Pneumatico vecchio (Gomma opaca) |
| 0 | 1 | Plastica lucida rossa (Palla da bowling) |
| 1 | 0 | Alluminio grezzo / Metallo sabbiato |
| 1 | 1 | Specchio cromato perfetto |

#### **C. Normal Map (Il "Fake" 3D)**

C'è uno slot chiamato "Normal Map".

* **Cos'è:** Qui si inseriscono texture speciali (di solito blu/viola).
* **Magia:** Serve a simulare piccoli dettagli (crepe, bulloni, graffi) **senza** aggiungere poligoni al modello 3D. Dice alla luce di rimbalzare "come se" ci fosse un buco, anche se la superficie è piatta.
* *Nota:* Se Unity ti dice "This texture is not marked as a Normal Map", clicca su "Fix Now".

#### **D. Emission (La Luce)**

* **Funzione:** Se lo attivi, l'oggetto emette luce propria (come uno schermo del PC, un neon o la spada laser).
* **Uso:** Scegli un colore e un'intensità. Non illumina gli altri oggetti (a meno che non usi luci "Baked"), ma rende l'oggetto visibile al buio.

### 3. Tiling & Offset (La Ripetizione)

Sotto le mappe principali, trovi questi valori numerici.

* **Tiling (X, Y):** Quante volte la texture si ripete sull'oggetto.
* *Esempio:* Se hai un pavimento enorme e la texture sgrana, metti Tiling X=10, Y=10 per ripeterla 10 volte e renderla più definita.

* **Offset:** Sposta la texture. Utile per scorrere l'acqua o i nastri trasportatori (animandolo via codice).
