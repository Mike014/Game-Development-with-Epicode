## FixedUpdate

### Cos’è

* Metodo legato al **motore fisico**
* Chiamato a intervalli **costanti**
* Non dipende dal frame rate

### Quando viene chiamato

* Può essere chiamato **0, 1 o più volte per frame**
* Dipende da `Time.fixedDeltaTime` (default 0.02 s)

### Quando usarlo

* **Fisica**
* **Rigidbody**
* Forze, velocità, `AddForce`, `MovePosition`

### Esempi

* Movimento con Rigidbody
* Salto fisico
* Applicazione di forze

## LateUpdate

### Cos’è

* Metodo chiamato **una volta per frame**
* Eseguito **dopo Update**

### Quando usarlo

* Operazioni che devono avvenire **dopo** che tutto si è mosso
* Tipico per:

  * **Camera follow**
  * aggiustamenti finali di posizione/rotazione
  * smooth camera

### Esempi

* Camera che segue il player
* Correzioni visive

## Ordine di esecuzione (semplificato)

```
FixedUpdate (0..n volte)
Update
LateUpdate
```

## Confronto rapido

| FixedUpdate | LateUpdate |
| ----------- | ---------- |
| Fisica      | Visuale    |
| Tempo fisso | Per frame  |
| Rigidbody   | Camera     |
| Forze       | Rifiniture |

## Regola d’oro

* **Fisica → FixedUpdate**
* **Input → Update**
* **Camera / rifiniture → LateUpdate**
