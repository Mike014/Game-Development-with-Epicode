# Collider vs Trigger: La Differenza Concettuale

In Unity, un Collider può esistere in due stati fondamentali, determinati dalla spunta **Is Trigger**. Questa scelta cambia radicalmente il modo in cui il motore fisico (PhysX) gestisce l'oggetto.

### 1. Il Collider Fisico (Is Trigger = OFF)

È la modalità "muro". L'oggetto possiede un corpo solido che occupa spazio nel mondo.

* **Comportamento:** Blocca il movimento. Calcola urti, rimbalzi e attrito.
* **Interazione:** Due oggetti non possono occupare lo stesso spazio; se si scontrano, si respingono.
* **Messaggio:** Invia l'evento `OnCollisionEnter`.
* **Esempi:** Muri, pavimenti, ostacoli, il corpo fisico di un personaggio.

### 2. Il Trigger (Is Trigger = ON)

È la modalità "sensore". L'oggetto diventa un volume invisibile e immateriale.

* **Comportamento:** Permette il passaggio. Il motore fisico ignora l'impatto ma rileva la sovrapposizione.
* **Interazione:** Funziona come un rilevatore di presenza. Registra quando qualcuno entra o esce dalla sua area, senza ostacolare il movimento.
* **Messaggio:** Invia l'evento `OnTriggerEnter`.
* **Esempi:** Checkpoint, zone di allerta, aree che attivano filmati o trappole.

### Uso avanzato: Compound Colliders (Collider Composti)

Molto spesso un singolo oggetto ha bisogno di entrambi i comportamenti. In Unity è possibile aggiungere **più Collider** allo stesso GameObject per gestire funzioni diverse:

* **Il Collider Solido:** (es. un *Capsule Collider*) serve a rendere l'oggetto "tangibile", impedendo che attraversi i muri o che il giocatore ci passi attraverso.
* **Il Trigger:** (es. uno *Sphere Collider* molto grande) serve a creare un'area di influenza attorno all'oggetto per "sentire" quando il giocatore si avvicina.

Questa tecnica permette di separare la **fisica del corpo** dalla **logica di rilevamento**.
