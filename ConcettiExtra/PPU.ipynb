# **Pixels Per Unit (PPU)**

Ecco la formula fondamentale e le sue derivazioni per gestire il rapporto tra pixel e spazio in Unity:

### 1. La formula base (Definizione di PPU)

Utilizzala per decidere quale valore inserire nell'Inspector dello Sprite:

$$\text{Unità Unity} = \frac{\text{Pixel}}{\text{PPU}}$$

### 2. Calcolo dello spazio (La più usata)

Utilizzala per sapere quanto spazio occuperà il tuo Sprite o per impostare la dimensione dei Collider:

$$\text{Unità Unity} = \frac{\text{Pixel}}{\text{PPU}}$$

### 3. Calcolo dell'area di gioco (Risoluzione e Camera)

Utilizzala per calcolare quante unità totali sono visibili a schermo data una risoluzione:

$$\text{Unità Totali (Orizzontali)} = \frac{\text{Risoluzione X}}{\text{PPU}}$$
$$\text{Unità Totali (Verticali)} = \frac{\text{Risoluzione Y}}{\text{PPU}}$$

### Il consiglio del Mentore

Ricorda che il **PPU** è come la "densità" del tuo oggetto.

* Se usi un **PPU alto** (es. 200), l'oggetto apparirà **più piccolo** in Unity (perché servono più pixel per riempire un'unità).
* Se usi un **PPU basso** (es. 32), l'oggetto apparirà **più grande** (tipico della Pixel Art, dove pochi pixel devono coprire molto spazio).
