## 1. Design Time (o Edit Time)

È la fase della **creazione intellettuale**. Qui il software esiste solo come **Codice Sorgente** (testo human-readable).

* **Cosa fa:** Il programmatore traduce la logica del problema in algoritmi usando un linguaggio (C#, C++, etc.).
* **Analisi Sezione per Sezione:** In questa fase l'IDE esegue un'analisi statica continua per segnalare errori di sintassi prima ancora di tentare la compilazione.
* **Pattern e Best Practices:** È qui che applichi i Design Pattern (es. Singleton, Observer) per garantire la manutenibilità futura.

## 2. Compile Time (Tempo di Compilazione)

Qui interviene il **Compilatore**. È la fase di traduzione "statica".

* **Cosa fa:** Trasforma il codice sorgente in **Codice Oggetto** (un formato intermedio o linguaggio macchina specifico per l'architettura).
* **Le sotto-fasi accademiche:**
1. **Analisi Lessicale:** Divide il testo in "parole" (token).
2. **Analisi Sintattica:** Verifica che la "grammatica" sia corretta (es. parentesi chiuse).
3. **Analisi Semantica:** Controlla la coerenza logica (es. non puoi sommare un numero a una parola).
4. **Generazione Codice:** Produce il file binario.



## 3. Link Time (Tempo di Collegamento)

Il compilatore crea file separati per ogni classe. Il **Linker** ha il compito di "cucirli" insieme.

* **Cosa fa:** Risolve i riferimenti esterni. Se la tua classe `Player` usa la libreria `UnityEngine.Physics`, il Linker trova dove risiede fisicamente quel codice e crea il collegamento.
* **Risultato:** Viene generato il file **Eseguibile finale** (.exe, .app).

## 4. Load Time (Tempo di Caricamento)

Questa è la fase "ponte" tra il disco rigido e la memoria RAM.

* **Cosa fa:** Il **Loader** (parte del Sistema Operativo) legge l'eseguibile dal disco, alloca lo spazio necessario nella RAM e carica le istruzioni.
* **Implicazioni:** Qui vengono assegnati gli indirizzi di memoria reali alle variabili che avevi definito solo "sulla carta".

## 5. Runtime (Tempo di Esecuzione)

Il processore (CPU) inizia a leggere le istruzioni dalla RAM una per una.

* **Cosa fa:** Esegue il ciclo **Fetch-Decode-Execute** (Preleva l'istruzione, decodificala, eseguila).
* **In Unity:** Qui entra in gioco il **Game Loop** (Update, FixedUpdate) e la gestione dinamica della memoria (Heap e Garbage Collector).

### Tabella Riassuntiva: Chi comanda in quale fase?

| Fase | Attore Principale | Output | Esempio di Errore |
| --- | --- | --- | --- |
| **Design** | Programmatore | `.cs` / `.cpp` (Testo) | Logica sbagliata |
| **Compile** | Compilatore | `.obj` / `.dll` (Binario) | Errore di sintassi |
| **Link** | Linker | `.exe` (Eseguibile) | Libreria mancante |
| **Load** | Loader | Immagine in RAM | Memoria RAM insufficiente |
| **Run** | CPU / VM | Risultato / Effetto | NullReferenceException |
