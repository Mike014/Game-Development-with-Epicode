### 1. Gradi vs Radianti: I Fondamenti

Immagina di voler descrivere una rotazione completa intorno a un punto.

* **I Gradi (°):** Sono l'unità "umana" per eccellenza. Dividono il cerchio in **360 parti uguali**. Sono facili da visualizzare (90° è un angolo retto, 180° è un'inversione di marcia).
* **I Radianti (rad):** Sono l'unità "matematica". Si basano sulla geometria intrinseca del cerchio. Un radiante è l'angolo formato quando la lunghezza dell'arco è uguale al raggio del cerchio stesso.

---

### 2. Schema di Conversione Rapida

Il punto di giunzione tra i due mondi è l'angolo piatto (), che equivale a  radianti (circa 3,14).

| Angolo | Gradi | Radianti | Nota Utile |
| --- | --- | --- | --- |
| **Giro Completo** |360°  | $2\pi$| Una rotazione totale. |
| **Angolo Piatto** |180° | $\pi$ | La base per tutte le conversioni. |
| **Angolo Retto** | 90° | $\pi/2$ | Fondamentale per le direzioni perpendicolari. |
| **Angolo Acuto** | 45°  | $\pi/4$ | La diagonale perfetta. |

---

### 3. Perché servono entrambi?

In Unity e nella programmazione in generale, noterai questa distinzione:

1. **Gradi:** Si usano nell'**Inspector** e per l'interfaccia utente perché sono intuitivi per noi.
2. **Radianti:** Si usano nelle **funzioni matematiche** (come `Mathf.Sin`, `Mathf.Cos`, `Mathf.Atan2`). I computer "pensano" in radianti perché rendono i calcoli trigonometrici molto più efficienti e lineari.
