## XR/VR in breve 

**VR (Virtual Reality)**
È un’esperienza **totalmente immersiva**: indossi un visore e l’ambiente “reale” viene **sostituito** da uno spazio digitale interattivo. Unity la descrive proprio così: contenuti immersivi che rimpiazzano l’ambiente dell’utente tramite HMD (head-mounted display). 

**XR (Extended Reality)**
È un **termine ombrello** che include più “realtà” immersive: **VR + AR + MR**. In pratica, XR = tutto lo spettro tra reale e virtuale. 

* **AR (Augmented Reality)**: sovrappone elementi digitali al mondo reale (informazioni, oggetti, UI).
* **MR (Mixed Reality)**: “mescola” reale e virtuale in modo più profondo (ancoraggi spaziali, interazioni più coerenti col mondo).
  (Le definizioni variano un po’ tra aziende/marketing, ma XR come ombrello è una costante.)
