### 1. Il Concetto di "Rig" (Lo Scheletro Digitale)

Immagina un modello 3D come una scultura di creta. Per farla muovere senza che si sgretoli, hai bisogno di un'armatura interna fatta di fil di ferro. In computer grafica, questa armatura è il **Rig**.

* **Cos'è:** È una gerarchia di "ossa" (chiamate *joints* o *bones*) collegate tra loro. Quando muovi un osso (es. l'avambraccio), la "pelle" del modello (la *mesh*) lo segue di conseguenza.
* **Perché è importante:** Senza un rig, un modello 3D è statico. Il rig definisce i punti di articolazione e i limiti di movimento.
* **Best Practice:** Nella scheda **Rig** di Unity, puoi scegliere quali ossa esporre nella *Hierarchy*. È fondamentale mostrare solo quelle necessarie (come le mani per impugnare un'arma) per non appesantire il motore di gioco con centinaia di oggetti inutilizzati.

### 2. Animation Type: Generic vs. Humanoid

Unity deve sapere come interpretare questo scheletro.

* **Generic:** Si usa per quadrupedi, mostri con molte braccia o oggetti meccanici. Unity vede solo una gerarchia di trasformazioni (muovi l'osso A, si sposta l'osso B).
* **Humanoid:** È un sistema speciale. Unity "sa" cos'è un essere umano. Se dici a Unity che il modello è Humanoid, lui cercherà di mappare le ossa del tuo file su una struttura standard (testa, braccia, gambe).

### 3. L'Avatar e il Retargeting

Questo è il cuore tecnologico dell'animazione in Unity.

* **L'Avatar:** Pensa all'Avatar come a un **dizionario bilingue**. Traduce i nomi delle ossa del tuo modello (magari chiamate "L_Arm_Joint") nel linguaggio standard di Unity ("LeftArm").
* **Retargeting:** Grazie a questo "dizionario", puoi prendere un'animazione di corsa registrata per un modello alto 2 metri e applicarla perfettamente a un nano o a un bambino. Unity adatta le proporzioni automaticamente perché entrambi usano lo standard Humanoid.

### 4. Animation Clips ed Events

Un singolo file FBX può contenere una lunga timeline con tutte le animazioni (camminata, corsa, salto).

* **Clips:** In Unity puoi "tagliare" questa timeline creando dei segmenti (es. dal frame 1 al 30 è "Idle", dal 31 al 60 è "Run").
* **Events:** Sono ponti tra animazione e codice. Se vuoi che un suono di "passo" venga riprodotto esattamente quando il piede tocca terra, inserisci un *Animation Event* in quel frame. Quell'evento chiamerà una funzione nel tuo script C#.

### 5. Gestione dei Materiali e Formati (FBX vs GLB)

Quando importi un modello, spesso appare grigio. Questo accade perché i materiali sono "dentro" il file originale e Unity non può modificarli.

* **Extraction:** Estrarre i materiali significa portarli fuori dal file e trasformarli in asset di Unity modificabili.
* **Il Formato GLB:** È come un "pacchetto pronto". A differenza dell'FBX (che è un formato proprietario di Autodesk molto flessibile ma complesso), il GLB (basato su glTF) è pensato per il web e lo scambio rapido: spesso mantiene i materiali già configurati, riducendo il lavoro manuale in Unity, a patto di avere i plugin giusti (*glTFast*).
