# **Sistema Tank Control e Combat System in Unity**

Le slide descrivono la costruzione di un **sistema di movimento e combattimento classico**, tipico dei primi survival horror, ponendo le basi sia **logiche** che **prestazionali**.

Il **Tank Control** è un sistema di input in cui il personaggio:

* si muove **in avanti/indietro** lungo la direzione in cui guarda
* **ruota su sé stesso** a sinistra/destra

Questo tipo di controllo, rigido e intenzionale, viene implementato in Unity tramite **Rigidbody**, separando chiaramente:

* **rotazione** (`MoveRotation`)
* **traslazione** (`MovePosition` con `transform.forward`)

Su questa base di movimento viene costruito un **sistema di armi**, che mostra diversi approcci al combat:

* armi **hit-scan** (raycast) per colpi istantanei
* proiettili fisici con **collisioni reali**
* armi ad area con **delay temporale ed esplosione**

L’**input del mouse** viene usato per attivare armi diverse, collegando ogni pulsante a un comportamento specifico, introducendo il concetto di **mapping input → azione**.

Per gestire eventi differiti (come l’esplosione di una granata), vengono presentati vari approcci al **timing**, dai più manuali (`Time.deltaTime`, `Time.time`) a soluzioni più immediate come `Invoke`, preparando il terreno a strumenti più avanzati come le **Coroutine**.

Dal punto di vista tecnico, le slide evidenziano problemi reali di **fisica e performance**:

* le collisioni discrete possono fallire con proiettili molto veloci
* per questo esistono modalità di **collision detection continua**, con diversi compromessi tra accuratezza e costo computazionale

Sul fronte delle esplosioni e dei controlli ad area, viene chiarita la differenza tra:

* metodi che **allocano memoria** (`OverlapSphere`)
* metodi **NonAlloc**, che riutilizzano array e riducono il Garbage Collector

Infine, tutto converge su un concetto chiave di produzione reale: **Object Pooling**.
Istanziare e distruggere continuamente oggetti è costoso; riutilizzarli tramite pool è una strategia fondamentale per:

* proiettili
* effetti
* esplosioni
* giochi action ad alto ritmo
