### **1. Gerarchia dei GameObjects**

* Trascinando un **GameObject** su un altro nella **Hierarchy**, il primo diventa *child* del secondo (*parent*).
* Tutti i figli ereditano le trasformazioni geometriche (posizione, rotazione, scala) del parent.
* Attenzione: se il parent ha una scala non uniforme (x, y, z diversi), possono verificarsi **distorsioni** sui figli.

---

### **2. Pivot dei GameObjects**

* Il **pivot** è il punto attorno a cui avvengono le trasformazioni (rotazioni, scalature).
* Nei modelli 3D, il pivot dipende da come è stato esportato (es. da Maya).
* In **Unity non è possibile cambiare il pivot**, ma si può creare un parent posizionato nel punto desiderato e muovere il child di conseguenza.

---

### **3. Organizzare i GameObjects**

* I **tag** sono etichette (una per oggetto) utili per catalogare i GameObjects e identificarli via codice (es. “Enemy”).
* I **layer** sono simili ai tag, ma servono anche per il rendering e la fisica.

  * Permettono di creare “piani di esistenza” separati (es. oggetti che non collidono tra loro o non sono visibili da certe camere).

---

### **4. Prefabs**

* Un **prefab** è un GameObject salvato come asset nella **Project Window**.
* Trascinando un GameObject dalla **Hierarchy** alla **Project Window** si crea un prefab.
* Le copie del prefab nella scena sono chiamate **istanze**.
* Se si modifica il prefab, le modifiche si applicano a **tutte le istanze**.

---

### **5. Overrides**

* Le modifiche fatte a una singola istanza di prefab si chiamano **overrides**.
* Non si possono eliminare i children originali, ma si possono aggiungere o modificare elementi.
* L’override “copre” le modifiche successive del prefab base.
* È possibile:

  * **Revert All** → annulla le modifiche e ripristina il prefab base.
  * **Apply All** → salva le modifiche e aggiorna tutte le istanze.

---

### **6. Prefab Mode**

* Si entra con doppio click sul prefab o cliccando la freccia accanto nella **Hierarchy**.
* Permette di modificare il prefab in un ambiente separato dalla scena.
* Tutte le modifiche fatte in questo modo si applicano direttamente all’asset e quindi a tutte le istanze.

---

### **7. Prefab Innestati**

* È possibile inserire prefab **dentro altri prefab**.
* Es.: prefab “Ramo” come child del prefab “Albero”.
* Modificando il prefab “Ramo”, si aggiornano automaticamente tutti gli alberi che lo contengono.

---

### **8. Prefab Variant**

* Creando una copia modificata di un prefab e trascinandola nella Project View, si può creare un **Prefab Variant**.
* Le varianti ereditano tutto dal prefab originale, ma possono avere modifiche personalizzate.
* Modificare il prefab originale aggiorna automaticamente anche le varianti, tranne dove ci sono override.
