# Ecosistema dello sviluppo

* **Tre livelli**

  * **Developers**: realizzano il gioco.
  * **Publishers**: finanziano e curano marketing; i dev possono essere interni/esterni.
  * **Distributors**: distribuzione fisica/digitale (es. Valve, Epic). Le grandi aziende possono coprire più ruoli.
* **Figure professionali**

  * **Game Designers** (regole/esperienza), **Programmers** (codice e tool), **2D/3D Artists** (grafica), **Composers/Sound Designers** (audio), **Project Managers** (coordinamento).
* **Linguaggi più usati**

  * **C++** (Unreal), **C#** (Unity), **Java** (Android), **JavaScript** (web), **Swift/Objective-C** (iOS).

# Game Engine

* Un gioco richiede motori per **logica**, **grafica**, **fisica** (e altro).
* Il **Game Engine** integra tutto e consente di concentrarsi sul design/giocabilità.
* I più diffusi: **Unreal Engine 5** e **Unity 3D**.

# Unity 3D: capacità e struttura progetto

* Supporta **2D/3D**, **VR/AR**, **simulazioni**, **rendering realtime**, **multiplatform**.
* Cartelle **indispensabili**: **Assets**, **Packages**, **ProjectSettings** (le altre sono rigenerabili).
* **Assets**: immagini (sprite/texture), modelli 3D (mesh/materiali/texture), **scripts C#** (componenti, editor personalizzati), audio/video.

# Scene, GameObjects e componenti

* **Scene**: sezioni del gioco (livelli, menu). Contengono tutto ciò che appare; gli elementi principali sono i **GameObjects**.
* **GameObjects**: personaggi/oggetti/luci ecc.; organizzabili in gerarchia **parent-child**. Ogni oggetto è definito da **componenti**.
* Componenti chiave:

  * **Transform** (sempre presente: posizione/rotazione/scala).
  * **Mesh Renderer** (visualizza il 3D).
  * **Collider** (collisioni).
  * **Rigidbody** (fisica: gravità/forze).
* Nuova scena tipica: **Camera**, **Directional Light**; in **URP** anche **Global Volume** (post-processing).
* **Camera**: proiezione **Perspective/Orthographic**, **FOV/Size**, **Viewport**, **Clipping Planes**.

# Panoramica Editor di Unity

**Project Window**

* Organizza/crea/cerca/filtra **Assets**.
* Shortcut: `Ctrl+Alt+C` copia percorso; `Ctrl+R` refresh; `Ctrl+D` duplica; `Canc` elimina; `Ctrl+F` cerca; doppio click apre asset/editor.

**Scene Window**

* Vista dello **sviluppatore** (non del giocatore); navigazione e cambio prospettiva.
* Navigazione rapida: frecce o **WASDQE** + mouse; `Shift` veloce; `Alt+sinistro` orbita; `Alt+centrale` pan; `Alt+destro` zoom; `F` centra su oggetto.

**Hierarchy Window**

* Elenco dei **GameObjects** della scena, gestione **parent-child**, creazione/ordinamento/rename tramite drag&drop.

**Inspector**

* Mostra e modifica **componenti** dell’oggetto/asset selezionato; **Add Component**; selezione multipla; per gli asset, valori e **import settings**.

**Toolbar**

* Strumenti: **Pan/Move/Rotate/Scale/Rect/Multiplo**; **Gizmo** (Pivot/Center, Local/Global, Snap); **Play/Pause/Step**; **Version Control/Account**; **Layers/Layout**.
* Shortcut essenziali: `Q` Pan, `W` Move, `E` Rotate, `R` Scale, `T` Rect, `Y` Multi, `Z` Pivot, `X` Local/Global, `V` Snap, `Ctrl+P` Play, `Ctrl+Shift+P` Pause, `Ctrl+Alt+P` Step.

**Game Window**

* Output del gioco (ciò che vede il giocatore) per test in **Play Mode**; controlli: **Aspect Ratio**, **Scale**, **Mute**, **Maximize**.
