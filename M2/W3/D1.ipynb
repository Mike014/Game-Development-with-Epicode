### **1️⃣ Funzioni – introduzione**

* Le **funzioni** sono procedure che possiamo richiamare più volte nel codice.
* Servono per:

  * **Risparmiare codice**, evitando di riscrivere la stessa procedura.
  * **Facilitare la manutenzione**, perché se serve modificare qualcosa, basta farlo una sola volta.
* Esempio base:

  ```csharp
  void Start() {
      Debug.Log("Stampiamo qualcosa.");
  }
  ```

---

### **2️⃣ Struttura di una funzione**

La formula generale è:

```csharp
tipo_di_dato NomeFunzione(parametri) {
    // corpo della funzione
}
```

* **Tipo di dato**: ciò che la funzione restituisce (es. `int`, `float`), oppure `void` se non restituisce nulla.
* **Nome della funzione**: identificatore scelto dallo sviluppatore.
* **Parametri (tra le parentesi)**: dati in input alla funzione (opzionali).
* **Corpo della funzione**: insieme delle istruzioni da eseguire.

---

### **3️⃣ Richiamare una funzione**

* Per **richiamare** una funzione, basta scrivere il suo nome e i parametri:

  ```csharp
  int risultato = Somma(10, 11, 12);
  ```
* Quando il programma la chiama:

  * L’esecuzione si ferma e la funzione viene eseguita.
  * Il risultato viene restituito nel punto della chiamata.
* Esempio:

  ```csharp
  int Somma(int a, int b, int c) {
      int somma = a + b + c;
      return somma;
  }
  ```
* `return` serve per **interrompere la funzione e restituire un valore**.

---

### **4️⃣ Overload di una funzione**

* Possiamo scrivere **più versioni** della stessa funzione (con lo stesso nome) ma con **parametri diversi**.
  → Questo si chiama **overloading**.
* Il compilatore capisce automaticamente quale versione usare in base ai parametri passati.
* Esempio:

  ```csharp
  int Somma(int a, int b) { return a + b; }
  int Somma(int a, int b, int c) { return a + b + c; }
  int Somma(int a, int b, int c, int d) { return a + b + c + d; }
  ```
