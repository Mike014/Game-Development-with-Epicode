### Wwise interface basics 

L’interfaccia di **Wwise** (di **Audiokinetic**) è divisa in **views**, ciascuna dedicata a un compito specifico. Le views vengono organizzate in **layouts**, pensati per facilitare diverse fasi di lavoro come authoring, integrazione e profiling.

Ogni layout include una **menu bar**, con i comandi generali del progetto, e una **toolbar**, che offre accesso rapido a strumenti essenziali come selezione di piattaforma e lingua, capture, connessione remota e ricerca.

---

### Wwise views

L’interfaccia di **Wwise** (di **Audiokinetic**) è composta da alcune **views chiave**, ognuna con uno scopo preciso:

* **Project Explorer**: centro di gestione del progetto

  * **Audio**: gerarchie Devices, Busses, Containers
  * **Events**: Action e Dialogue Events
  * **SoundBanks**: gestione delle bank
  * **Game Syncs**: States, Switches, RTPC, Triggers
  * **ShareSets**: effetti e attenuazioni condivise
  * **Sessions**: Soundcaster
  * **Queries**: ricerche strutturate

* **Contextual Help**: aiuto e spiegazioni delle proprietà selezionate

* **Object Tabs**: editor contestuali degli oggetti selezionati

* **Transport Control**: play / stop / pause degli oggetti

* **Meter View**: livelli audio (Peak, True Peak, RMS, e 3D Meter)

### Using text boxes, sliders & controls 

In **Wwise** (di **Audiokinetic**), le proprietà si modificano tramite **campi standardizzati**:

* **Text boxes**: inserimento diretto dei valori

  * *Ctrl+click* → reset al valore di default
  * *Click & hold* → **super slider** per regolazioni fini
  * *Shift+drag* → variazioni più precise

* **Sliders**: regolazione continua dei valori

  * lo slider mostra la posizione del valore rispetto al range
  * il range visibile può essere solo una parte del range reale

* **Lists**: selezione di opzioni predefinite

  * **drop-down list** per scegliere valori
  * **shortcut list (>>) ** per azioni o opzioni rapide

* **Faders**: slider verticali (es. Volume)

  * simulano mixer reali
  * *Shift+click* per micro-aggiustamenti
  * *Ctrl+click* → reset

* **Tables**: editing e gestione di grandi quantità di dati

  * colonne configurabili, ordinabili e ridimensionabili
  * ricerca rapida (*Ctrl+F3*)
  * editing diretto dei valori

* **Undo / Redo**:

  * *Ctrl+Z* / *Ctrl+Y*
  * fino a **200 azioni**
