# Riassunto Strutturato Completo

## 📦 **Fondamenti (Sezioni 1-3)**
**Unity = Editor a Componenti**: assembli comportamenti attaccando script agli oggetti
- **Script C#** → devono ereditare da `MonoBehaviour` per diventare componenti
- **Lifecycle**: `Start()` (setup) → `Update()` (loop continuo)
- **Inspector**: `public` mostra variabili, `private` nasconde

## 🎮 **Controllo e Dati (4-5)**
**Control Flow**: `if/else`, operatori logici (`&&`, `||`)
**Collezioni**:
- Array → fissi, veloci
- List → flessibili, `Add()`/`Remove()`
- Dictionary → coppie chiave-valore

**Design**: GDD, TDD, One-Page per pianificare prima di costruire

## 🏃 **Movimento e Fisica (6-7)**
**3 Strategie**:
1. `Transform` → teleport semplice
2. `Rigidbody` → fisica realistica (**MAI mischiare con Transform!**)
3. `CharacterController` → soluzione pronta

**Collisioni**:
- Collider → blocca
- Trigger (`OnTriggerEnter`) → rileva senza bloccare

**Animazioni**: Animation Clip (dati) → Controller (macchina a stati) → Animator (componente)

## 🎯 **Meccaniche Core (8-9)**
- **Salto**: `AddForce(Vector3.up * force, ForceMode.Impulse)`
- **Layer Masks**: identificano cosa si sta toccando (es. "Ground")
- **Sparo**: `Instantiate(prefab)` + `Destroy(obj, delay)` = "Fire and Forget"

## 🎛️ **Sistemi Globali (10-12)**
**Game Manager**: controlla logica globale (punteggio, vita)
- `private` variabili + `public` Properties → incapsulamento
- `SetActive(bool)` → mostra/nasconde UI
- `Time.timeScale = 0` → pausa
- `SceneManager.LoadScene()` → restart

**AI**: NavMesh (mappa camminabile) + NavMesh Agent (movimento autonomo) + A* (pathfinding)

## ⚔️ **Sistema di Combattimento Completo**
**Architettura Modulare**:
- Ogni GameObject gestisce il **proprio comportamento** (danni inflitti/subiti, reazioni)
- **Separazione delle responsabilità**: 
  - Nemico → pathfinding + attacco a distanza
  - Giocatore → input + vita
  - Game Manager → stato globale + GUI sync

**Interazioni Nemico-Giocatore**:
- Nemico usa NavMesh per **muoversi** → quando distanza < threshold → passa a **modalità attacco**
- Danni bidirezionali + aggiornamento UI in tempo reale
- Pattern: **evento locale** (collisione) → **reazione individuale** → **notifica Manager**

## 🧹 **Principi di Programmazione Procedurale**
**DRY (Don't Repeat Yourself)**:
- Codice ripetuto → astrarre in **metodi dedicati**
- Esempio: invece di 5 blocchi identici per danni, un metodo `TakeDamage(int amount)`
- Risultato: codice più **leggibile**, **manutenibile**, **testabile**
